# Spam Text Message Classification

### Import the necessary libraries

```python
import pandas as pd
import nltk 
import re 
import string
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score
from sklearn.ensemble import RandomForestClassifier
```

### Import the dataset

```python
dataset = pd.read_csv("data/spam.csv", encoding='latin-1')
dataset
```

```python
# Remove unneccesary columns
dataset = dataset[["v1","v2"]]

#renaming of columns
dataset.rename(columns={"v1":"labels","v2":"text"}, inplace = True)
dataset
```

```python
dataset.shape
```

### Clean the data

```python
stopwords = nltk.corpus.stopwords.words('english')

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text

```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(dataset['text'])

X_features = pd.DataFrame(X_tfidf.toarray())
X_features.head()
```

### Model with RandomForestClassifier

```python
print(RandomForestClassifier())
```

```python
# split data into training and testing 
training_text, testing_text, training_labels, testing_labels = train_test_split(X_features, dataset["labels"], test_size = 0.2)
```

```python
# fit random forest model 
rf = RandomForestClassifier()
rf_model = rf.fit(training_text, training_labels)
```

```python
# Make predictions on the test set using the fit model
y_pred = rf_model.predict(testing_text)
```

```python
# Evalute model predictions using precision and recall
precision = precision_score(testing_labels, y_pred, pos_label='spam')
recall = recall_score(testing_labels, y_pred, pos_label='spam')
print('Precision: {} / Recall: {}'.format(round(precision, 3), round(recall, 3)))
```
